```python
# Merge all year data together 
```

```python
data_dir = "/Users/jiagu/Desktop/Data_Project/tennis-ml-project/data/raw"
files = sorted([
    f for f in os.listdir(data_dir)])

col_sets = {}

for file in files: 
    path = os.path.join(data_dir, file)
    try: 
        df = pd.read_csv(path, nrows=1)
        col_sets[file] = set(df.columns)
    except Exception as e: 
        print(f"error reading {file}: {e}")

unique_col_sets = set(frozenset(cols) for cols in col_sets.values())
print(f"Unique column sets: {len(unique_col_sets)} across {len(files)} files.\n")
```

```python
complete_df = []

for file in files: 
    path = os.path.join(data_dir, file)
    print(f"loading: {file}")
    df = pd.read_csv(path)
    complete_df.append(df)

all_matches = pd.concat(complete_df, ignore_index = True )
```

```python
# Code	Meaning
# DA	Direct Acceptance (based on ranking)
# WC	Wild Card — invited by the tournament despite low or no ranking
# Q	Qualifier — advanced from the qualifying rounds
# LL	Lucky Loser — lost in qualifying, but got in due to another player’s withdrawal
# SE	Special Exempt — couldn’t play qualifiers due to playing in previous week
# PR	Protected Ranking — entered based on pre-injury ranking
# ALT	Alternate — filled in for a last-minute withdrawal
# SR	Special Ranking (same as Protected Ranking)
# ITF	Entered based on ITF World Tennis Ranking

all_matches['winner_entry'].value_counts()
```

```python
all_matches['score'].value_counts()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Correlations in the numeric attributes 

numeric_df = all_matches.select_dtypes(include=['float64', 'int64'])
numeric_df = numeric_df.drop(columns=['tourney_id', 'match_num', 'winner_id', 'loser_id'], errors='ignore')

sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title("Numeric values' correlation")
plt.show()
```

```python
# Relationship between rank difference and match duration (There are outliers)

all_matches['rank_diff'] = all_matches['winner_rank'] - all_matches['loser_rank'] 
sns.scatterplot(x='rank_diff', y = 'minutes', data = all_matches)
plt.title("Ranking Difference vs. Match Duration")
```

```python
# Winner Rank by Surface

sns.boxplot(x='surface', y='winner_rank', data = all_matches)
plt.title('Winner rank by Surface')
```

```python
# serve stats compared between winner and loser 

sns.histplot(all_matches['w_ace'], bins=40, color='green', label='Winner Aces', kde=True)
sns.histplot(all_matches['l_ace'], bins=40, color='red', label='Loser Aces', kde=True)
plt.legend()
plt.title("Aces by Winners vs. Losers")
```
